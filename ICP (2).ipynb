## Preprocess data

### Subtask:
Prepare the data for modeling. This may involve handling missing values, encoding categorical variables, and scaling numerical features.


**Reasoning**:
Identify categorical features, apply one-hot encoding, separate the target variable, scale numerical features, and concatenate the results.



```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

# Separate features and target
X = df.drop('target', axis=1)
y = df['target']

# Identify categorical features (assuming columns with '_cat' or '_bin' are categorical)
categorical_features = [col for col in X.columns if '_cat' in col or '_bin' in col]
numerical_features = [col for col in X.columns if col not in categorical_features and col != 'id']

# Create transformers for numerical and categorical features
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Create a column transformer to apply different transformations to different columns
# Explicitly drop 'id' column
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='drop' # Drop any columns not specified in transformers (like 'id')
)

# Apply transformations
X_processed = preprocessor.fit_transform(X)

# Convert to DataFrame (optional, but useful for inspection)
# Get feature names after one-hot encoding
categorical_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
all_feature_names = numerical_features + list(categorical_feature_names)


# Ensure X_processed is a numpy array before creating DataFrame
if not isinstance(X_processed, np.ndarray):
    X_processed = X_processed.toarray()


X_processed_df = pd.DataFrame(X_processed, columns=all_feature_names)

display(X_processed_df.head())
display(y.head())
```

## Split data

### Subtask:
Split the data into training and testing sets to evaluate the model's performance on unseen data.


**Reasoning**:
Split the processed data into training and testing sets and print their shapes.



```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed_df, y, test_size=0.2, random_state=42)

# Print the shapes of the resulting sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Build a model

Train a predictive model (e.g., a classification model) on the training data to predict whether a customer will buy the product.


```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Build the model
# Due to the imbalanced nature of the target variable,
# using a tree-based model like Gradient Boosting is a good approach.
# We can start with default parameters and tune later if needed.
model = GradientBoostingClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Calculate AUC
auc = roc_auc_score(y_test, y_pred_proba)
print(f"\nAUC: {auc:.4f}")
```

## Interpret model
Interpret the model's results to understand which features are most important in predicting customer behavior.


Examine the feature importances from the trained model to identify which features contribute most to the predictions.

```python
# Get feature importances
feature_importances = model.feature_importances_

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({
    'feature': X_train.columns,
    'importance': feature_importances
})

# Sort by importance
feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)

# Display top 10 important features
display(feature_importance_df.head(10))

# Optional: Visualize feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance_df.head(10))
plt.title('Top 10 Most Important Features')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
